```python
%matplotlib inline
import numpy as np
from sklearn import svm
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
X = np.array([[-1, -1], [-2, -1], [-3, -2], [1, 1], [2, 1], [3, 2]])
plt.scatter(X[:,0], X[:,1])
```

```python
pca = PCA(n_components=2)
pca.fit(X)
print(pca.explained_variance_ratio_)
print(pca.singular_values_)
```

```python
cumulative_explained = [0]
for i in range(len(pca.explained_variance_ratio_)):
    cumulative_explained.append(sum(pca.explained_variance_ratio_[:i+1]))
plt.plot(cumulative_explained)
```

```python
print(pca.components_)
print(pca.explained_variance_)
print(np.var(X[:,0]))
print(np.var(X[:,1]))
cov = np.cov(X, rowvar=False)
print(cov)
eig = np.linalg.eig(cov)
print(eig)


x = np.arange(-3, 4)
y0 = x * pca.components_[0][1] / pca.components_[0][0]
y1 = x * pca.components_[1][1] / pca.components_[1][0]
plt.scatter(X[:,0], X[:,1])
plt.plot(x, y0)
plt.plot(x, y1)
```
